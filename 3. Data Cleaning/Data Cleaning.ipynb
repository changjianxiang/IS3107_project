```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("merged_carparks.csv")
```

```python
df
```

```python
df.dropna(inplace=True)
#df.reset_index(inplace=True)
```

```python
#Dropping uncessary column
df.columns.str.match("Unnamed")
df = df.loc[:,~df.columns.str.match("Unnamed")]
df.drop(columns=['tempIndex','ingestionTime'],inplace=True)
df.reset_index(inplace=True,drop=True)
df
```

```python
#Extract day,hour,weekend,weekday,filled proportion from time column
df['FilledProportion'] = (df['totalCapacity'] - df['lotsAvailable']) / df['totalCapacity']

df['ingestionDateTime'] = pd.to_datetime(df['ingestionDateTime'])
df['hour'] = df['ingestionDateTime'].apply(lambda x : x.hour)
#0 = Monday 6 = Sunday
df['weekday'] = df['ingestionDateTime'].apply(lambda x : x.weekday())
df['weekend'] = np.where(df['weekday'].isin([5,6]),1,0)
df['day'] = df['ingestionDateTime'].apply(lambda x : x.day)
```

```python
#df['startTime'].unique()
df['endTime'].unique()
```

```python
#columns
df.columns
```

```python
#Extract hour from time
def splitter(x):
    item = x.split('.')
    hour = item[0]
    appendage  = item[1][3:]
    if appendage == "PM":
        hour = int(hour) + 12
    return int(hour)
df['strippedEndTime'] = df['endTime'].apply(lambda x : splitter(x))
df['strippedStartTime'] = df['startTime'].apply(lambda x : splitter(x))
print(df['strippedEndTime'].unique())
print(df['strippedStartTime'].unique())
```

```python
#Remove rows based on rates applicability
index = []

for i in range(0,len(df)):
    if (df.iloc[i]['hour'] <= df.iloc[i]['strippedEndTime']):
        if (df.iloc[i]['hour'] >= df.iloc[i]['strippedStartTime']):
            pass
        else:
            index.append(i)
    else:
        index.append(i)
df.drop(index,axis=0,inplace=True)
df.reset_index(inplace=True,drop=True)
df
```

```python
#Get actual rates
df['rate'] = ""
df
for i in range(0,len(df)):
    if df.iloc[i]['weekend'] == 0:
        df.at[i,'rate'] = df.iloc[i]['weekdayRate']
    else:
        if df.iloc[i]['weekday'] == 5:
            df.at[i,'rate'] = df.iloc[i]['satdayRate']
        else:
            df.at[i,'rate'] = df.iloc[i]['sunPHRate']
            
df
```

```python
#Trim the remaining uncessary column
df.drop(columns=['lotTypeR'],inplace=True)
df
```

```python
df.to_csv("processed_merged_carparks.csv", encoding='utf-8', index=False)
```
